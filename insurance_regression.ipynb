```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Load data
data = pd.read_csv('insurance.csv')
```

```python

# Inspect the data
print(data.head())
```

```python
# Identify categorical and numeric columns
categorical_cols = ['sex', 'smoker', 'region']
numeric_cols = ['age', 'bmi', 'children']
```

```python
# Prepare data for linear regression
X = data[['age']]  # Independent variable (feature)
y = data['charges']  # Dependent variable (target)
```

```python
# Plot scatter plot of age vs charges
plt.figure(figsize=(8,6))
plt.scatter(X, y, color='blue', alpha=0.5)
plt.title('Scatter plot of Age vs Insurance Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
#Create and fit the Linear Regression model
model = LinearRegression()
model.fit(X, y)
 
```

```python
# Predictions
y_pred = model.predict(X)
```

```python
# Plot scatter plot with best-fit regression line
plt.figure(figsize=(8,6))
plt.scatter(X, y, color='blue', alpha=0.5, label='Actual data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Best-fit line')
plt.title('Age vs Insurance Charges with Linear Regression Line')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.legend()
plt.show()
```

```python
# Model evaluation
from sklearn.metrics import mean_squared_error, r2_score  # ensure this is present
mse = mean_squared_error(y, y_pred)
rmse = mse ** 0.5  # Manual square root of MSE
r2 = r2_score(y, y_pred)

# Get model coefficients
intercept = model.intercept_
coef = model.coef_[0]

# Print results
print(f"Regression Equation: charges = {coef:.2f} * age + {intercept:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R^2 Score: {r2:.3f}")
```

**Conclusion**
The linear regression model relating age to insurance charges is expressed as:
charges= 257.72 x age+ 3165.89

This equation suggests that for each additional year of age, insurance charges increase on average by approximately R257.72.
However, the models 2 score is 0.089, indicating that only about 8.9% of the variability in insurance charges can be explained by age alone. This suggests that age, while having some effect, is not a strong predictor of charges by itself.

The relatively high RMSE (~R11,552) also shows that the model's predictions have a considerable error margin, which means other factors likely play a significant role in determining insurance costs.

Overall, while age is positively correlated with insurance charges, a more complex model including additional variables would be needed for better prediction accuracy.
